### Scrapping quotes with author detail:

```python
import requests
import pandas as pd
from tqdm import tqdm 
from bs4 import BeautifulSoup
```

```python
link = "https://quotes.toscrape.com/"
res = requests.get(link)
soup = BeautifulSoup(res.text,'html.parser')
```

```python
soup.find_all('div',class_='quote')
```

```python
data=[]

for sp in soup.find_all('div',class_='quote'):
    
    quote     = sp.find('span',class_='text').text[1:-1]
    author    = sp.find('small',class_="author").text
    author_id = sp.find('a').get('href')
    author_link= link+author_id
    tags = []
    for tag in  sp.find_all('a',class_="tag"):
        tags.append(tag.text)
    tags =  ','.join(tags)
    data.append([quote,author,author_id,tags,author_link])
```

```python
df=pd.DataFrame(data,columns=['quote','author','author_id','tags','author_link'])
```

```python
df
```

```python
df.to_csv('quotes.csv',index=False)
```

### Scrapping author info:

```python
import pandas as pd 
from bs4 import BeautifulSoup
from tqdm import tqdm
import requests
```

```python
df2 = pd.read_csv("quotes.csv")
```

```python
df2.head()
```

```python
del df2['quote']
del df2['author_id']
del df2['tags']
```

```python
df2 = df2.drop_duplicates()
```

```python
len(df2.values)
```

```python
df2
```

```python
author_info = []
paths = []


for author in tqdm(df2.values):
    name = 'Author' + author[0] + '.txt'
    link = author[1]
    res = requests.get(link)
    soup = BeautifulSoup(res.text, 'html.parser')
    
    author_details = soup.find('div', class_='author-details')
    
    author_info.append(author_details)
    paths.append(name)
    
    if author_details:
        text = author_details.get_text(strip=True)  # Extract the text from the element
        
        # Write the information to a text file
        fd = open(name, 'w', encoding='utf-8')
        fd.write(text)
        fd.close()
        
```
