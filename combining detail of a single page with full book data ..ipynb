```python
import pandas as pd 
```

```python
df_1 = pd.read_csv('books.csv')
df_2= pd.read_csv('data.csv')
```

```python
df_2.head()
```

## Creating New dataframe:


```python
df = pd.DataFrame()
```

```python
df['title']              = df_1['title']
df['upc']                = df_2['upc']
df['category']           = df_2['category']
df['price_ex']           = df_2['price_ex']
df['price_in']           = df_2['price_in']
df['tax']                = df_2['tax']
df['book_rating']        = df_1['book_rating']
df['Number_of_reviews']  = df_2['Number_of_reviews']
df['Availability']       = df_2['Availability']
df['stock_availability'] = df_1['stock_availability']
df['img_link']           = df_1['img_link']
df['book_link']          = df_1['book_link']
```
